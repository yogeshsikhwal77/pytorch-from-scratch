```python
import torch
```

## Ex1

```python
x = torch.tensor(3.0,requires_grad=True)
```

```python
y = x**2
```

```python
print(x)
print(y)
```

```python
y.backward()
```

```python
x.grad
```

## Ex2

```python
x = torch.tensor(3.0,requires_grad=True)
```

```python
y = x**2
z = torch.sin(y)
```

```python
print(x)
print(y)
print(z)
```

```python
z.backward()
```

```python
x.grad
```

```python
y.grad
```
