```python
import torch
```

```python
f = torch.randint(size=(2,3),low=0,high=10)
g = torch.randint(size=(3,2),low=0,high=10)
print(f)
print(g)
```

```python
torch.matmul(f,g) #matrix multification
```

```python
v1 = torch.tensor([1,2])
v2 = torch.tensor([3,4])
```

```python
torch.dot(v1,v2)
```
