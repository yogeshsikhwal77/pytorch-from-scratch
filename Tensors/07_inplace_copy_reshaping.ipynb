## Copy a tensor

```python
a = torch.rand(2,3)
a
```

```python
b = a
```

```python
b
```

```python
a[0][0] = 0
```

```python
a
```

```python
b  # change come in b also
# means changes in both original or copy 
```

```python
print(id(a))
print(id(b))
```

```python
b = a.clone()
```

```python
print(id(a))
print(id(b)) # different location so changes not in both if i change any one
```

## Tensor in gpu

```python
# creating a new tensor on gpu
torch.rand((2,3),device=device)
```
